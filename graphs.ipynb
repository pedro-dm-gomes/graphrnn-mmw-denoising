```python

import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

import matplotlib.pyplot as plt
import numpy as np
import json
import os
import pandas as pd

base_path = './experiments/'

# load 'history.csv' file from experiments/ subdirectory
def load_history(experiment_dir):
    history_path = os.path.join(experiment_dir, 'history0.csv')
    return pd.read_csv(history_path)

# load subdirectories from base_path in a list
def load_experiments(base_path):
    experiments = []
    for experiment in sorted(os.listdir(base_path)):
        if os.path.isdir(os.path.join(base_path, experiment)):
            experiments.append(experiment)
    return experiments

```

```python
experiments = load_experiments(base_path)
history = load_history(os.path.join(base_path, experiments[-1]))
```

```python
# plot training and validation loss in a (16,9) figure
plt.figure(figsize=(16,9))
plt.plot(history['loss'], label='loss')
plt.plot(history['val_loss'], label='val_loss')
plt.legend()
```

```python
# plot training and validation accuracy in a (16,9) figure
plt.figure(figsize=(16,9))
plt.plot(history['sparse_categorical_accuracy'], label='acc')
plt.plot(history['val_sparse_categorical_accuracy'], label='val_acc')
plt.legend()
```
